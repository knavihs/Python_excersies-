```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
import pandas as pd
```

```python

arry=np.random.randint(0,3000,size=3000)
arry
```

```python
df = pd.DataFrame(arry)
df.head(10)
```

```python

import matplotlib.pyplot as plt
wss = []
k = range(1,15)
for i in k:
    km = KMeans(n_clusters= i,n_init=25,max_iter=30)
    km = km.fit(df)
    wss.append(km.inertia_)
```

```python
plt.plot(k,wss,'bx-')
plt.xlabel('No. of Clusters')
plt.ylabel('Within sum of squares')
plt.title('Optimal no. of clusters')
plt.show()
```

```python
# as per the elbow suggests , optimal number of clusters should be 3
km = KMeans(n_clusters=4,n_init=25,max_iter=30).fit(df)
wss_current = km.inertia_
print(wss_current,' ',wss[3] )
```

```python
idx=df.index
```

```python
plt.scatter(df,idx)
plt.show
```

```python
y_means = km.predict(df)
plt.scatter(df,idx,c=y_means,s=50,cmap='viridis')
centers = km.cluster_centers_
print(centers)
plt.scatter(centers, centers, c='black', s=200, alpha=0.5);
```
